```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_excel('ogrenci.xlsx')
df.head()
```

```python
df.loc[df['Ebeveyn Egitim Seviyesi']=='Lise Mezunu','Ebeveyn Egitim Seviyesi']= 1
df.loc[df['Ebeveyn Egitim Seviyesi']=='Onlisans Mezunu','Ebeveyn Egitim Seviyesi']= 2
df.loc[df['Ebeveyn Egitim Seviyesi']=='Lisans Mezunu','Ebeveyn Egitim Seviyesi']= 3
df.loc[df['Ebeveyn Egitim Seviyesi']=='Yuksek Lisans Mezunu','Ebeveyn Egitim Seviyesi']= 4
df.loc[df['Cinsiyet']=='Kadin','Cinsiyet']= 0
df.loc[df['Cinsiyet']=='Erkek','Cinsiyet']= 1
df.loc[df['Okul Yemekhanesi']=='Yok','Okul Yemekhanesi']= 0
df.loc[df['Okul Yemekhanesi']=='Var','Okul Yemekhanesi']=1
df.loc[df['Ozel Ders']=='Yok','Ozel Ders']=0
df.loc[df['Ozel Ders']=='Var','Ozel Ders']=1
```

```python
df.head()
```

```python
x=df.drop(['Matematik','Okuma','Yazma'],axis=1)
y=df[['Matematik','Okuma','Yazma']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
dt=DecisionTreeRegressor(random_state=42)
dt2=DecisionTreeRegressor(random_state=42)
dt3=DecisionTreeRegressor(random_state=42)
dt_fit_mat=dt.fit(X_train,y_train['Matematik'])
dt_fit_oku=dt2.fit(X_train,y_train['Okuma'])
dt_fit_yaz=dt3.fit(X_train,y_train['Yazma'])


feature_mat=dt_fit_mat.feature_importances_
feature_oku=dt_fit_oku.feature_importances_
feature_yaz=dt_fit_yaz.feature_importances_

feature_names=x.columns
df_content=pd.DataFrame({
    "Özellik":feature_names,
    "Matematik":feature_mat,
    "Okuma":feature_oku,
    "Yazma":feature_yaz
})
```

```python
df_content
```

```python
y_pred=dt.predict(X_test)
y_pred2=dt2.predict(X_test)
y_pred3=dt3.predict(X_test)

accuracy_mat=dt.score(X_test,y_test['Matematik'])
accuracy_oku=dt2.score(X_test,y_test['Okuma'])
accuracy_yaz=dt3.score(X_test,y_test['Yazma'])

print('Matematik Doğruluk Oranı:',accuracy_mat)
print('Okuma Doğruluk Oranı:',accuracy_oku)
print('Yazma Doğruluk Oranı:',accuracy_yaz)
```

```python
df = pd.read_excel('ogrenci.xlsx')
df.head()

df.loc[df['Ebeveyn Egitim Seviyesi']=='Lise Mezunu','Ebeveyn Egitim Seviyesi']= 1
df.loc[df['Ebeveyn Egitim Seviyesi']=='Onlisans Mezunu','Ebeveyn Egitim Seviyesi']= 2
df.loc[df['Ebeveyn Egitim Seviyesi']=='Lisans Mezunu','Ebeveyn Egitim Seviyesi']= 3
df.loc[df['Ebeveyn Egitim Seviyesi']=='Yuksek Lisans Mezunu','Ebeveyn Egitim Seviyesi']= 4
df.loc[df['Cinsiyet']=='Kadin','Cinsiyet']= 0
df.loc[df['Cinsiyet']=='Erkek','Cinsiyet']= 1
df.loc[df['Okul Yemekhanesi']=='Yok','Okul Yemekhanesi']= 0
df.loc[df['Okul Yemekhanesi']=='Var','Okul Yemekhanesi']=1
df.loc[df['Ozel Ders']=='Yok','Ozel Ders']=0
df.loc[df['Ozel Ders']=='Var','Ozel Ders']=1
```

```python
x=df.drop(['Matematik','Okuma','Yazma'],axis=1)
y=df[['Matematik','Okuma','Yazma']]
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

lr=LinearRegression()
feature=lr.fit(X_train,y_train['Matematik'])
lr2=LinearRegression()
feature2=lr2.fit(X_train,y_train['Okuma'])
lr3=LinearRegression()
feature3=lr3.fit(X_train,y_train['Yazma'])
y_pred=lr.predict(X_test)
y_pred2=lr2.predict(X_test)
y_pred3=lr3.predict(X_test)
accuracy=lr.score(X_test,y_test['Matematik'])
accuracy2=lr2.score(X_test,y_test['Okuma'])
accuracy3=lr3.score(X_test,y_test['Yazma'])
print("Matematik Doğruluk Oranı:",accuracy)
print("Okuma Doğruluk Oranı:",accuracy2)
print("Yazma Doğruluk Oranı:",accuracy3)

print("Matematik Katsayı:",feature.coef_)
print("Okuma Katsayı:",feature2.coef_)
print("Yazma Katsayı:",feature3.coef_)




```

```python
# Okuma becerisinin yazma ve matematik üzerindeki etkisi nedir?
x=df[['Okuma']]
y=df[['Matematik','Yazma']]
lr4=LinearRegression()
feature4=lr4.fit(x,y['Matematik'])
lr5=LinearRegression()
feature5=lr5.fit(x,y['Yazma'])
print("Matematik Katsayı:",feature4.coef_)
print("Yazma Katsayı:",feature5.coef_ )
```

```python
#Özel ders almanın genel başarıya etkisi
df['Genel Basari']= (df['Matematik']+df['Okuma']+df['Yazma'])/3
x=df[['Ozel Ders']]
y=df['Genel Basari']

lr=LinearRegression()
fit=lr.fit(x,y)

print("Katsayı",fit.coef_)

print("Sabit",fit.intercept_)
```
